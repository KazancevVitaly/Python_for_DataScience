### Задача 1
***Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.***

При увеличении Precision мы уменьшаем количество ложно-положительных угадываний,
т.е. увеличиваем количество ложно-отрицательных угадываний, т.е. уменьшаем Recall.
И наоборот, при увеличении recall мы увеличиваем количество ложно-положительных угадываний,
уменьшая при это количество ложно-отрицательных, тем самым уменьшая Precision.

#### Примеры когда лучше максимизировать Precison:
* 1 Задача оттока клиентов (за нулевую гипотезу берем, что клиент лоялен к организации).
* 2 Если банк заинтересован в получении прибыли, то решение о том давать кредит(0) или не давать(1) принимать исходя из precision

#### Примеры когда лучше максимизировать Recall:
* 1 Болеет человек COVD-19 или нет (за нулевую гипотезу берем, что болеет).
* 2 Определение наличия опасных веществ/предметов в багаже (за нулевую гипотезу бреме что в багаже имеется опасное вещество/предмет).

### Задача 2
***Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?***
Увеличивая Precision (далее P) мы уменьшаем Recall (далее R), и наоборот, поэтому простое арефмитическое среднее между P и R не даст объективного представления о качестве модели. F-мера дает представление как о Precision так и о Recall, выдавая как результат гармоническое среднее между величинами P и R. В тоже время позволяет(в некторых своих вариациях) акцентировать внимание на той величине, максимизация которой для нас более актуальна при решении конкретной задачи.
